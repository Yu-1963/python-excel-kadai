```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
from openpyxl.styles import Font

# Excelファイル名のリスト
file_names = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

# 各Excelファイルを読み込み、指定したシートのデータを格納するリストを作成
dataframes = [
    pd.read_excel(file_name, sheet_name='Sheet1')
    for file_name in file_names
]

# リスト内の全てのDataFrameを連結
all_data = pd.concat(dataframes, ignore_index=True)

# '商品'と'売上年'ごとに'金額（千円）'を集計
summary = all_data.groupby(['商品', '売上年'])['金額（千円）'].sum().reset_index()

# 集計結果を新しいExcelファイルに書き込み
summary_file = '売上集計表.xlsx'
summary.to_excel(summary_file, index=False)

# openpyxlでExcelファイルを開き、ヘッダーの書式を変更
book = load_workbook(summary_file)
sheet = book.active
for cell in sheet[1]:
    cell.font = Font(bold=True)

# 変更を保存
book.save(summary_file)
```
